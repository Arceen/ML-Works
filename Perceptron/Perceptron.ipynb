```python
# Creating perceptron class
# Only the sigmoid activation function has been implemented for forward feed and backpropagation
import numpy as np
class Perceptron:
    def __init__(self, num_inputs, activation, learning_rate):
        self.activation = activation
        self.learning_rate = learning_rate
        self.W = list(np.random.rand(num_inputs+1))
    
    
    def train_epochs(self, epochs, x_train, y_train):
        for i in range(epochs):
            self.true = 0
            self.all = 0
            print(f"Epoch {i+1} :", end = ' ')
            for j in range(len(y_train)):
                self.train(x_train[j], y_train[j])
            print(f"Accuracy: {(self.true/self.all)*100}%")
            
        
        print(f"Final Weights: {self.W}")
    
    def delta_weights(self, y_pred, y):
        if self.activation == 'sigmoid':
            return (1 - y_pred) * y_pred * (y_pred - y)
    
    def predict(self, pred):
        if self.activation == 'sigmoid':
            return self.sigmoid_activation(pred)
        
    def sigmoid_activation(self, x):
        return 1/(1+np.exp(-x))
    
    def train(self, x, y):
        X = x.copy()
        X.append(1)
        X = np.array(X)
        weights = np.array(self.W)
        
        y_pred = self.predict(weights @ X)
        
        for i in range(len(self.W)):
            self.W[i] -= self.learning_rate * self.delta_weights(y_pred, y) *  X[i]
        
        # Checking Accuracy with confusion matrix definition
        if self.checkeval(y_pred) == y:
            self.true += 1
        self.all += 1 

    
    def checkeval(self, x):
        return 1 if x >= 0.5 else 0
    
    def evaluate(self, X):
        X.append(1)
        X = np.array(X)
        weights = np.array(self.W)
        return self.predict((X @ weights))
    
    def classifyeval(self, x):
        return self.checkeval(self.evaluate(x))
```

```python
epochs = 2000

input_size = 2

#0.1 ~ 0.01 usually works fast in most cases here
learning_rate = 0.01

AND_INPUTS = [[0, 0], [0, 1], [1, 0], [1, 1]]
AND_OUTPUTS = [0, 0, 0, 1]

OR_INPUTS = [[0, 0], [0, 1], [1, 0], [1, 1]]
OR_OUTPUTS = [0, 1, 1, 1]

# Input size need to be changed to 1
NOT_INPUTS= [[0], [1]]
NOT_OUTPUTS = [1, 0]

# Does not work for XOR
XOR_INPUTS = [[0, 0], [0, 1], [1, 0], [1, 1]]
XOR_OUTPUTS = [0, 1, 1, 0]
```

```python
input_size = 2
p = Perceptron(input_size, 'sigmoid', learning_rate)
p.train_epochs(epochs, OR_INPUTS, OR_OUTPUTS)
```
