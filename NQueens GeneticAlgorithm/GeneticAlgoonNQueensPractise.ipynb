```python
def attacks(queens):
    rows = [0 for i in range(len(queens))]
    main_diagonals, aux_diagonals = [[0 for i in range(len(queens)*2-1)] for i in range(2)]
    attacks = 0
    for i in range(len(queens)): 
        r = queens[i] - 1
        c = i
        attacks += rows[r]                                                                                                                                                                      
        attacks += main_diagonals[r-c+len(queens)-1]
        attacks += aux_diagonals[r+c]
        rows[r]+=1;
        main_diagonals[r-c+len(queens)-1] += 1
        aux_diagonals[r+c]+= 1;
    
    return attacks
```

```python
#Delete bottom half of all states(less performing)
def remove_weak_states(Queens): 
    half = len(Queens)
    rmv = min(half, 10)
    weak_states = []
    for i, k in enumerate(Queens):
        if i >= rmv or Queens[k] <= 15:
            weak_states.append(k)

    for state in weak_states:
        del Queens[state]

Queens
```

```python
def get_curr_states(Queens):
    l = []
    for i in Queens.keys():
        l.append([*i])
    
    return l
```

```python
def crossover(s1, s2):
    crossover_point = rand.randrange(1, N-1)
    l1 = s1[:crossover_point]+s2[crossover_point:]
    l2 = s2[:crossover_point]+s1[crossover_point:]
    return l1, l2
    
```

```python
def next_generation(current_generation, Queens):
    parentN = len(current_generation)
    for i in range(len(current_generation)):
        for j in range(i+1, len(current_generation)):
            children = crossover(current_generation[i], current_generation[j])
            for child in children:
                if tuple(child) not in Queens:
                    Queens.update({tuple(child) : fitness_function(N, attacks(child))})
                    
                # Putting a chance on Mutation
                mutation = rand.randrange(1,30)
                if mutation == 1:
                    mutationspot = rand.randrange(0,N)
                    mutationvalue = rand.randrange(1, N+1)
                    child[mutationspot] = mutationvalue
                    if tuple(child) not in Queens:
                        Queens.update({tuple(child) : fitness_function(N, attacks(child))})
                        print("-----\n MUTATION OCCURED \n----")
                    
    
    Queens = OrderedDict(sorted(Queens.items(), key = lambda x: x[1], reverse=True))
    return Queens
```
