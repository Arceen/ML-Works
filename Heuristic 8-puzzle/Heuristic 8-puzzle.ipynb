```python
def pos_cst(cst):
    t = copy.deepcopy(cst)
    return {tuple(v):k for k, v in t.items()}
def curr_pos(cst):
    return cst[' ']
def findbypos(pos, cst):
#     print("\nfindbypos: ")
#     print(cst)
#     print(pos_cst(cst))
#     print(pos)
    return pos_cst(cst)[tuple(pos)]
```

<h1>Manhattan distance</h1>

```python
def h2(cstate, gstate):
    return sum( [abs(cstate[i][0]-gstate[i][0])+abs(cstate[i][1]-gstate[i][1]) for i in range(1,9)])
```

```python
def h1(cstate, gstate):
    return  len([i for i in range(1,9) if gstate[i] != cstate[i]])
```

```python
#Make a move. Change current state. Update the info of 2 states in the current state and return the state
def move(cs, poscurr, posnext):
#     print("\nMOVE: ")
    x = copy.deepcopy(cs)
    x[findbypos(posnext, x)] = poscurr
    x[' '] = posnext
#     print(x)
    return x
```

```python
#return list of available moves based on the current position of the move_directions
def move_directions(cs):
    d = copy.deepcopy(cs)
    pos = curr_pos(d)
#     print("\nmove directions: ")
#     print(pos)
#     print(d)
    av_mv = []
    if pos[0] > 0:
        #can move up
#         av_mv.append(("UP", [pos[0]-1, pos[1]])
#         print("UP")
        av_mv.append(move(d, pos, [pos[0]-1, pos[1]]))
#         print(d)
    if pos[1] > 0:
#         print("LEFT")
        #can move left
#         av_mv.append(("LEFT", [pos[0], pos[1]-1]))
        av_mv.append(move(d, pos, [pos[0], pos[1]-1]))
#         print(d)
    if pos[0] < r-1:
#         print("DOWN")
        #can move down
#         av_mv.append(("DOWN", [pos[0]+1, pos[1]]))
        av_mv.append(move(d, pos, [pos[0]+1, pos[1]]))
#         print(d)
    if pos[1] < c-1:
#         print("RIGHT")
        #can move right
#         av_mv.append(("RIGHT", [pos[0], pos[1]+1]))
        av_mv.append(move(d, pos, [pos[0], pos[1]+1]))
#         print(d)
    
    return av_mv
    
```

```python
def heuristic(cs, gs):
    return h1(cs, gs) + h2(cs, gs)
```

```python
# Given a list l, need to sort by heuristic
#  is in [state, heuristic]
def sort_heuristiclist(l):
    return l.sort(key=lambda x: x[2])
```

    <li>Initialize empty moves_list from current state</li>

    <li>Make that move</li>
    <li>Check to see if that move resulted in reaching goal state</li>

```python

def search_best_moves(cs, gs):
    #cs and gs are DICTIONARY
    #copying the current move to store and modify
    #dictionary of previous move of a list
    #parent_map is going to map from encoded curr state dictionary to previous encoded state dictionary
    #items in move_list map from state to heuristic
    parent = {0: None}
    min_h = 100
    min_s = None
    dic_list = [cs]
    dic_counter = 0
    visited_states = []
    s = cs.copy()
    moves_list = []
    moves_list.append((dic_list.index(cs), heuristic(cs, gs), 0))
    found = False
    while not found and moves_list:
        #popping the smallest heuristic move
        curr_state = moves_list.pop(0)
        if curr_state[1] < min_h:
            min_h = curr_state[1]
            min_s = dic_list[curr_state[0]]
        curr_g = curr_state[2]
        curr_state = curr_state[0]
        
        visited_states.append(curr_state)
        #finding the available child states
        for mv in move_directions(dic_list[curr_state]):
            if mv == gs:
                #backtrack and print all
                print("Goal states Found")
                goal_moves = [mv]
                d = curr_state
                while  d != None:
                    goal_moves.append(dic_list[d])
                    print(dic_list[d])
                    d = parent[d]
                
                return goal_moves
                while goal_moves:
                    t = goal_moves.pop(-1)
                    print(t)
                found = True
            
            if mv not in dic_list:
                dic_counter += 1
                dic_list.append(mv)
                cc = dic_counter
            else:
                cc = dic_list.index(mv)
            
            if cc not in visited_states:
                #calculating the heuristics for this state
                h = heuristic(mv, gs)
                f = curr_g+h+1
                moves_list.append((cc, h, f))
                parent[cc] = curr_state
        
        sort_heuristiclist(moves_list)
        print(max(visited_states))
        print("Heuristics: ")
        print(min_h)
        print("Good State: ")
        print(min_s)
        
    
        
```
